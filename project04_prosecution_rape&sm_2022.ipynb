```python
import pandas as pd
df = pd.read_csv('prosecution_monthly.csv')
df
```

```python
# Filter rows where "Unnamed: 3" contains "2022年" and "Unnamed: 5" contains "強制性交"
result_df = df[df['Unnamed: 3'].str.contains('2022年', na=False) & df['Unnamed: 5'].str.contains('強制性交', na=False)]

# Display the extracted rows
print(result_df)
```

```python
# List of columns to keep
columns_to_keep = ['Unnamed: 3', 'Unnamed: 5', 'Unnamed: 8', 'Unnamed: 12']

# Drop the unnecessary columns
new_df = result_df[columns_to_keep]

# Display the updated DataFrame
print(new_df)
```

```python
new_df
```

```python
# Convert the values in the "Unnamed: 8" column to numeric (remove commas and convert to integers)
new_df['Unnamed: 8'] = new_df['Unnamed: 8'].str.replace(',', '').astype(int)

# Sum up all the numbers in the "Unnamed: 8" column
rape_total_sum = new_df['Unnamed: 8'].sum()

# Display the sum
print(rape_total_sum)
```

```python
new_df['Unnamed: 12'] = new_df['Unnamed: 12'].str.replace(',', '').astype(int)

nonrape_total_sum = new_df['Unnamed: 12'].sum()
print(nonrape_total_sum)
```
